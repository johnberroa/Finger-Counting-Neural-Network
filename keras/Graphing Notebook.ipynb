```python
import numpy as np
import matplotlib.pyplot as plt
```

# SIMPLE CNN

```python
tacc = np.load('Training Data/Simple/t-acc.npy')
tloss = np.load('Training Data/Simple/t-loss.npy')
vacc = np.load('Training Data/Simple/v-acc.npy')
vloss = np.load('Training Data/Simple/v-loss.npy')
```

```python
print("LAST T-ACC:", tacc[-1], "MAX T-ACC:", max(tacc))
print("LAST V-ACC:", vacc[-1], "MAX V-ACC:", max(vacc))
```

```python
fig, ax = plt.subplots(1,2, figsize=(20, 10))

# plt.suptitle("Simple CNN Performance")
plt.rc('axes', titlesize=20) 
plt.rc('axes', labelsize=20)

ax[0].plot(tacc[:200], c='b', label="Training")
ax[0].plot(vacc[:200], c='orange', label="Validation")
ax[0].axhline(.2, ls='--', c='r', label="Chance")
ax[0].set_xlabel("Training step")
ax[0].set_ylabel("Accuracy")
ax[0].set_title("Accuracy (max on val: {:.2f})".format(max(vacc)))
ax[0].legend()

ax[1].plot(tloss[:200], c='b', label="Training")
ax[1].plot(vloss[:200], c='orange', label="Validation")
ax[1].set_xlabel("Training step")
ax[1].set_ylabel("Loss")
ax[1].set_title("Loss (min on val: {:.2f})".format(min(vloss)))
ax[1].legend()

# plt.tight_layout()
# plt.savefig('simple_cnn.png', bbox_inches='tight')
plt.show()
```
